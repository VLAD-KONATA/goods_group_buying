```python
import os

path=rf"./旧排发表/切煤表/"

lists=os.listdir(path)
for foldername in lists:
    if '.files' in foldername:
        os.rename(os.path.join(path,foldername),os.path.join(path,foldername.split('.files')[0]))
        print('{name}→{name2}',foldername,foldername.split('.files')[0])

lists2=os.listdir(path)
for filename in lists2:
    if '.htm' in filename:
        os.remove(os.path.join(path,filename))
        print('remove {name}',filename)
```
